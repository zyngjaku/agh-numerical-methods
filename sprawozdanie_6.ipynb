```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

L=4.8
A=0.1 
T=0.2

def f(x):
    return A*np.sin(2*math.pi*x/T)

def f_prim(x):
    return 2*math.pi*A/T*np.cos(2*math.pi/T*x)

def dlugoscKrzywejBezCalki(x):
    return np.sqrt(1+f_prim(x)**2)
    
print("Metoda simpsona ", simpson(dlugoscKrzywejBezCalki, 0, L, 100))    
print("Metoda trapezów: ", trapezy(dlugoscKrzywejBezCalki, 0, L, 100))
print("Metoda romberga: ", integrate.romberg(dlugoscKrzywejBezCalki, 0, L))

pod=[]
sim=[]
tra=[]

for i in range(1, 100+1):
    pod.append(i)
    sim.append(simpson(dlugoscKrzywejBezCalki, 0, L, i))
    tra.append(trapezy(dlugoscKrzywejBezCalki, 0, L, i))
    
plt.figure(figsize=(20, 10))
plt.plot(pod, sim, 'g', label='metoda trapezów')
plt.plot(pod, tra, 'b', label='metoda simpsona')
plt.grid()
plt.xlabel('ilość podprzedziałów')
plt.ylabel('wartość obliczona')
plt.legend()
plt.show()
```

Metorda Romberga znacząco odbiega od wyników dla złożonych metod Simpsona i trapezów. Czym większa liczba podprzedziałów tym bardziej dokładniejszy wynik zwracają złożone metody Simpsona i trapezów (z małymi odchyleniami, czym więcej podprzedziałów tym mniej odchyleń).
