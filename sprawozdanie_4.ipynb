```python
import matplotlib.pyplot as plt
import numpy as np

def interpolacja(x, y):
    assert len(x)==len(y)!=0
    def f(n):
        s=0
        i=0
        while i<len(x):
            p=1
            j=0
            while j<len(x):
                if(j!=i):
                    p=p*(n-x[j])/(x[i]-x[j])
                j+=1
            s += y[i]*p
            i+=1
        return s
    return f
```

```python
import numpy as np
import math
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

def f(x): 
    return 1/(25*x**2 +1)

x = np.arange(-2,2,0.05)

x_rownoodlegle = np.linspace(-2, 2, 21)
x_czybyszew = np.array([2*math.cos(math.pi * (2 * k + 1) / (2 * 21)) for k in range(1, 21)])

wielomianowe_rownoodlegle = np.poly1d(np.polyfit(x_rownoodlegle, f(x_rownoodlegle), deg=(20)))
lagrange_czebyszew  = np.poly1d(np.polyfit(x_czybyszew, f(x_czybyszew), deg=(20)))
sklejane_rownoodlegle = interp1d(x_rownoodlegle, f(x_rownoodlegle), 3, fill_value='extrapolate')
sklejane_czebyszew  = interp1d(x_czybyszew, f(x_czybyszew), 3, fill_value='extrapolate')



plot1 = plt.figure(figsize=(10, 5), dpi=80, facecolor='w', edgecolor='k')

plt.plot(x, lagrange_rownoodlegle(x), label='Wzy Lagrage')
plt.plot(x, lagrange_czebyszew(x), label='Wzy Czebyczewa')

plt.scatter(x_rownoodlegle, wielomian_rownoodlegle(x_rownoodlegle))
plt.scatter(x_czybyszew, wielomian_czebyszew(x_czybyszew))

plt.title('Interpolacja Lagrange')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.ylim(-0.1,1.1)
plt.show



plot2 = plt.figure(figsize=(10, 5), dpi=80, facecolor='w', edgecolor='k')

plt.plot(x, sklejane_rownoodlegle(x), label='Wzy Rownoodlege')
plt.plot(x, sklejane_czebyszew(x), label='Wzy Czebyczewa')

plt.scatter(x_rownoodlegle, wielomian_rownoodlegle(x_rownoodlegle))
plt.scatter(x_czybyszew, wielomian_czebyszew(x_czybyszew))

plt.title('Interpolacja funkcji sklejanych')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.ylim(-0.1,1.1)
plt.show()
```
